```python
import pandas as pd
import json
from lib.Sahara.attribution import safety_head_attribution
import os
```

```python
splits = {'330k_train': 'round0/330k/train.jsonl.xz', '330k_test': 'round0/330k/test.jsonl.xz', '30k_train': 'round0/30k/train.jsonl.gz', '30k_test': 'round0/30k/test.jsonl.gz'}
df = pd.read_json("hf://datasets/PKU-Alignment/BeaverTails/" + splits["330k_train"], lines=True)
```

# drug_abuse,weapons,banned_substance

```python
val_drug = pd.read_csv("../exp_data/Beaver_samples/drug_abuse,weapons,banned_substance.csv")
```

```python
drug_train = df[df.apply(lambda x: x["category"]["drug_abuse,weapons,banned_substance"], axis = 1)]
```

```python
drug_train = drug_train[drug_train["prompt"].isin(val_drug["prompt"].tolist()).__invert__()]
```

```python
drug_train.drop_duplicates(subset="prompt").rename(columns={"prompt":"input"})["input"].to_csv("exp_data/unused_beaver/drug_abuse,weapons,banned_substance.csv")
```

# privacy_violation

```python
val_privacy = pd.read_csv("../exp_data/Beaver_samples/privacy_violation.csv")
```

```python
privacy_train = df[df.apply(lambda x: x["category"]["privacy_violation"], axis = 1)]
```

```python
privacy_train = privacy_train[privacy_train["prompt"].isin(val_privacy["prompt"].tolist()).__invert__()]
```

```python
privacy_train.drop_duplicates(subset="prompt").rename(columns={"prompt":"input"})["input"].to_csv("exp_data/unused_beaver/privacy_violation.csv")
```
